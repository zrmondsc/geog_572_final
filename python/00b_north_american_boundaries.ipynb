```python
import geopandas as gpd
```

```python
# input shapefile
in_path = r"C:\Users\Zachary\phd_classes\geog572\final\data\external\boundaries_p_2021_v3.shp"

# output geojson
out_path = r"C:\Users\Zachary\phd_classes\geog572\final\data\processed\boundaries_p_2021_v3.geojson"
```

```python
# read shapefile
gdf = gpd.read_file(in_path)
```

```python
tolerance = 1000  # meters
gdf["geometry"] = gdf.geometry.simplify(tolerance=tolerance, preserve_topology=True)
```

```python
gdf = gdf.copy().to_crs(4326)
```

```python
gdf.head(2)
```

```python
# save
gdf.to_file(out_path, driver="GeoJSON")
```
